```python
# Sample 2D array
arr = [
    ["apple", "banana", "cherry"],
    ["dog", "", "elephant"],
    ["fish", "-n", "goat"],
    ["hat", None, "ice"]
]

# Scan the array
for i in range(len(arr)):
    for j in range(len(arr[i])):
        value = arr[i][j]

        # Check if null/empty
        if value is None or value == "":
            print(f"Null/Empty found at [{i}][{j}]")

        # Check if termination string
        elif value == "-n":
            print(f"Termination string '-n' found at [{i}][{j}]")

        else:
            print(f"Value at [{i}][{j}] = {value}")
```

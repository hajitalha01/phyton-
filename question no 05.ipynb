Write a program that:
• Reads a text file named "data.txt".
• Counts how many times each unique word appears using a dictionary.
• Ignore punctuation and case sensitivity.
• Print the five most frequent words along with their counts.

```python
import string

try:
    with open("/content/17.txt", "r") as file:
        text = file.read()
except FileNotFoundError:
    print("Error: 'data.txt' file not found!")
    exit()

text = text.lower()
for punc in string.punctuation:
    text = text.replace(punc, "")

words = text.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

print("Top 5 Most Frequent Words:\n")
for word, count in sorted_words[:5]:
    print(f"{word}: {count}")
```

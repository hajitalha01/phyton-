```python
import pygame
import random

# Initialize
pygame.init()

# Screen
width = 400
height = 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("🚗 Car Racing Game")

# Colors
white = (255, 255, 255)
black = (0, 0, 0)
red = (200, 0, 0)

# Player car
car_width = 50
car_height = 80
car_x = width // 2 - car_width // 2
car_y = height - 100
car_speed = 5

# Enemy car
enemy_width = 50
enemy_height = 80
enemy_x = random.randint(0, width - enemy_width)
enemy_y = -100
enemy_speed = 5

clock = pygame.time.Clock()
score = 0

running = True

while running:
    screen.fill(white)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Controls
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and car_x > 0:
        car_x -= car_speed
    if keys[pygame.K_RIGHT] and car_x < width - car_width:
        car_x += car_speed

    # Enemy movement
    enemy_y += enemy_speed
    if enemy_y > height:
        enemy_y = -100
        enemy_x = random.randint(0, width - enemy_width)
        score += 1
        enemy_speed += 0.5  # speed increase 🔥

    # Collision
    if (car_x < enemy_x + enemy_width and
        car_x + car_width > enemy_x and
        car_y < enemy_y + enemy_height and
        car_y + car_height > enemy_y):
        print("💀 Game Over! Score:", score)
        running = False

    # Draw cars
    pygame.draw.rect(screen, black, (car_x, car_y, car_width, car_height))
    pygame.draw.rect(screen, red, (enemy_x, enemy_y, enemy_width, enemy_height))

    # Score display
    font = pygame.font.SysFont(None, 30)
    text = font.render("Score: " + str(score), True, black)
    screen.blit(text, (10, 10))

    pygame.display.update()
    clock.tick(60)

pygame.quit()
```

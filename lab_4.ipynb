## 3. Function with Parameters

```python

def add_numbers(a, b):
    c = a+b
    return c

print("Sum:", add_numbers(10, 5))
```

```python

def add_numbers(a, b):
    return a + b

print("Sum:", add_numbers(10, 5))
```

```python
def print_uppercase(text):
    return text.upper()

print(print_uppercase("python"))
```

```python
def sum_list(numbers):
    return sum(numbers)

nums = [1, 2, 3, 4, 5]
print("Sum of list:", sum_list(nums))
```

```python
def print_tuple_elements(items):
    for item in items:
        print(item)

fruits = ("apple", "banana", "cherry")
print_tuple_elements(fruits)
```

```python
def print_tuple_elements(items):
    for item in items:
        print(item, end=" ")

fruits = ("apple", "banana", "cherry")
print_tuple_elements(fruits)
```

```python
def print_student_details(student):
    for key, value in student.items():
        print(key, ":", value)

student_info = {"name": "Sara", "age": 21, "grade": "A"}
print_student_details(student_info)
```

```python
def unique_items(myset):
    print("Unique values are:")
    for item in myset:
        print(item)

unique_items({10, 20, 30, 20, 10})
```

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print("Factorial of 5:", factorial(5))
```

## Default Parameter

```python
def greet_user(name="Guest"):
    print("Hello,", name)

greet_user("Ali")
greet_user()  # uses default
```

```python
def add_numbers(a=10, b=66):
    return a + b

print("Sum:", add_numbers())
print("Sum:", add_numbers(10, 5))
```

## Function with Multiple Parameters & Default

```python
#Power Function
def power(base, exp=2):
    return base ** exp

print(power(5))      # uses default exponent 2 → 25
print(power(5, 3))   # overrides → 125
```

```python
#Area of Rectangle (Default Width)
def area(length, width=10):
    return length * width

print(area(5))        # width defaults to 10 → 50
print(area(5, 20))    # override → 100
```

```python
#Student Information
def student(name, grade="Not Assigned", age=18):
    print("Name:", name)
    print("Grade:", grade)
    print("Age:", age)

student("Sara")
student("Ali", grade="A")
student("John", grade="B", age=20)
```

```python
#Shopping Cart Example
def add_to_cart(item, quantity=1, price=100):
    total = quantity * price
    print(quantity, item, "added. Total =", total)

add_to_cart("Apple")            # default qty = 1, price = 100
add_to_cart("Mango", 5)         # qty = 5, price default
add_to_cart("Banana", 10, 20)   # qty = 10, price = 20
```

```python

def calculate_price(price, tax=0.05, discount=0.1):
    final_price = price + (price * tax) - (price * discount)
    return final_price

print("Price with default tax and discount:", calculate_price(100))
print("Price with custom values:", calculate_price(100, tax=0.08, discount=0.2))
```

```python
#Salary Calculator
def salary(base, bonus=1000, tax=0.1):
    return base + bonus - (base * tax)

print("Default:", salary(5000))          # bonus=1000, tax=0.1
print("Custom bonus:", salary(5000, 2000))
print("Custom bonus and tax:", salary(5000, 2000, 0.2))
```

```python
#API Request Simulation
def api_request(url, method="GET", timeout=30):
    print("Requesting:", url)
    print("Method:", method)
    print("Timeout:", timeout, "seconds")

api_request("http://example.com")
api_request("http://example.com", method="POST", timeout=10)
```

## 3. Modules (Importing and Using)

## 2.1 Built-in Modules¶

Modules in Python are pre-defined packages of functions and variables that you can import and use without rewriting the logic. For example, the math module provides mathematical functions

```python
import math
result = math.sqrt(16)
print(result) 
```

### 2.2 Importing Specific Functions
Python allows importing only specific functions or variables from a module to reduce memory usage and make code more readable.

```python
from math import pi
print(pi)
```

```python
import math
print(math.sqrt(16))  # square root
print(math.pi)         # value of pi
print(math.factorial(6))
```

```python
import random
print(random.randint(1, 10))   # random integer
print(random.choice(['apple', 'banana', 'cherry']))  # random choice from list
```

Write a recursive function named is_palindrome(s) that checks whether the given string s is a
palindrome or not.
Requirements:
• The function should ignore spaces, punctuation, and case sensitivity.
• It should handle both even-length and odd-length strings (e.g., "madam", "abba", "A man a
plan a canal Panama").
• If the input string contains any numeric digits, the function should ignore them during
comparison but not remove them from the original string.
• The function should return True or False accordingly, and also print "Palindrome" or "Not
Palindrome" before returning.

```python
import string

def is_palindrome(s):
    s = s.lower()
    cleaned = ""
    for ch in s:
        if ch.isalpha():
            cleaned += ch
    def check(word):

        if len(word) <= 1:
            return True

        if word[0] == word[-1]:
            return check(word[1:-1])
        else:
            return False

    result = check(cleaned)
    if result:
        print("Palindrome")
    else:
        print("Not Palindrome")
    return result

text = "A man a plan a canal Panama"
print(is_palindrome(text))
```

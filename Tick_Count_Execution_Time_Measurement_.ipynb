```python
import random
import time
import matplotlib.pyplot as plt

# -------------------------
# Step 1: Random Generation
# -------------------------
start1 = time.time()
print("Random Generation Start Time:", start1)

numbers = [random.randint(1, 1000) for _ in range(1000)]

end1 = time.time()
print("Random Generation End Time:", end1)
print("Random Generation Total Time:", (end1 - start1) * 1000, "ms")
print("-----------------------------------")


# -------------------------
# Step 2: Sorting
# -------------------------
start2 = time.time()
print("Sorting Start Time:", start2)

sorted_numbers = sorted(numbers)

end2 = time.time()
print("Sorting End Time:", end2)
print("Sorting Total Time:", (end2 - start2) * 1000, "ms")
print("-----------------------------------")


# -------------------------
# Step 3: Squaring
# -------------------------
start3 = time.time()
print("Squaring Start Time:", start3)

squared_numbers = [x**2 for x in sorted_numbers]

end3 = time.time()
print("Squaring End Time:", end3)
print("Squaring Total Time:", (end3 - start3) * 1000, "ms")
print("-----------------------------------")


# -------------------------
# Step 4: Plot Sorted Numbers
# -------------------------
start4 = time.time()
print("Sorted Plot Start Time:", start4)

plt.figure()
plt.plot(sorted_numbers)
plt.title("Sorted Numbers")
plt.show()

end4 = time.time()
print("Sorted Plot End Time:", end4)
print("Sorted Plot Total Time:", (end4 - start4) * 1000, "ms")
print("-----------------------------------")


# -------------------------
# Step 5: Plot Squared Numbers
# -------------------------
start5 = time.time()
print("Squared Plot Start Time:", start5)

plt.figure()
plt.plot(squared_numbers)
plt.title("Squared Numbers")
plt.show()

end5 = time.time()
print("Squared Plot End Time:", end5)
print("Squared Plot Total Time:", (end5 - start5) * 1000, "ms")
print("-----------------------------------")
```

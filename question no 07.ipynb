Create a program that:
â€¢ Continuously asks the user to enter numbers until they type 'stop'.
â€¢ Uses a while loop to calculate and display the average of all even numbers entered.
â€¢ If no even number is entered, display an appropriate message.

```python
even_numbers = []

print("Enter numbers one by one. Type 'stop' to finish.\n")

while True:
    value = input("Enter a number (or 'stop'): ").strip()

    if value.lower() == "stop":
        break

    if value.isdigit() or (value.startswith('-') and value[1:].isdigit()):
        num = int(value)
        if num % 2 == 0:
            even_numbers.append(num)
    else:
        print("Please enter a valid number!")
5737
if len(even_numbers) > 0:
    average = sum(even_numbers) / len(even_numbers)
    print("\nEven numbers entered:", even_numbers)
    print("Average of even numbers:", average)
else:
    print("\nNo even numbers were entered.")
```

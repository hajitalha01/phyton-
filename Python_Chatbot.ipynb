```python
print("Chatbot: Hello! I am your chatbot.")
print("Type 'bye' to exit.\n")

while True:
    user = input("You: ").lower()

    if user == "bye":
        print("Chatbot: Goodbye! 👋")
        break

    elif "hello" in user or "hi" in user:
        print("Chatbot: Hi there!")

    elif "how are you" in user:
        print("Chatbot: I'm fine, thanks!")

    elif "name" in user:
        print("Chatbot: I am a simple Python chatbot.")

    elif "help" in user:
        print("Chatbot: I can answer basic questions like hello, name, etc.")

    elif "time" in user:
        from datetime import datetime
        print("Chatbot:", datetime.now().strftime("%H:%M:%S"))

    elif "date" in user:
        from datetime import datetime
        print("Chatbot:", datetime.now().strftime("%Y-%m-%d"))

    else:
        print("Chatbot: Sorry, I don't understand that.")
```

```python
import random

print("🎮 Welcome Haji Talha to the Guessing Game!")
print("You have 5 attempts to guess the correct number (1-50)")

secret_number = random.randint(1, 50)
attempts = 5
score = 100

while attempts > 0:
    guess = int(input("Enter your guess: "))

    if guess == secret_number:
        print("🎉 Congratulations Haji Talha! You guessed it right!")
        print("Your Score:", score)
        break
    elif guess > secret_number:
        print("Too High!")
    else:
        print("Too Low!")

    attempts -= 1
    score -= 20
    print("Attempts left:", attempts)
    print("-------------------")

if attempts == 0:
    print("😢 Game Over!")
    print("The correct number was:", secret_number)
```

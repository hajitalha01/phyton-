```python
# ==============================
# Feature Selection Examples
# RFE + Forward Selection + Backward Selection
# ==============================

import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import (
    RFE,
    SequentialFeatureSelector
)

# ------------------------------
# Dataset
# ------------------------------

data = {
    'Study_Hours': [8,7,2,3,9,1,6,2],
    'Attendance': [90,85,40,50,95,30,80,35],
    'Mobile_Usage': [9,8,7,10,6,11,8,10],
    'Result': [1,1,0,0,1,0,1,0]
}

df = pd.DataFrame(data)

# Features (X)
X = df[['Study_Hours', 'Attendance', 'Mobile_Usage']]

# Target (y)
y = df['Result']

# ------------------------------
# Model
# ------------------------------

model = LogisticRegression()

# =====================================================
# 1. Recursive Feature Elimination (RFE)
# =====================================================

print("\n===== RFE =====")

rfe = RFE(
    estimator=model,
    n_features_to_select=2
)

rfe.fit(X, y)

print("Selected Features:")
print(X.columns[rfe.support_])

# =====================================================
# 2. Forward Selection
# =====================================================

print("\n===== Forward Selection =====")

forward = SequentialFeatureSelector(
    model,
    n_features_to_select=2,
    direction='forward'
)

forward.fit(X, y)

print("Selected Features:")
print(X.columns[forward.get_support()])

# =====================================================
# 3. Backward Selection
# =====================================================

print("\n===== Backward Selection =====")

backward = SequentialFeatureSelector(
    model,
    n_features_to_select=2,
    direction='backward'
)

backward.fit(X, y)

print("Selected Features:")
print(X.columns[backward.get_support()])
```

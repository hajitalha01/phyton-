Create a text file named students.txt that stores the records of 10 students.
Each line in the file should contain the following information, separated by commas:
Name, Age, Grade
Example:

Ali, 20, 88
Write a program to:
• Read the file line by line.
• Use an if-else condition to categorize each student as “Above Average” (grade ≥ 80) or
“Below Average”.
• Store results in a dictionary and save them to a new file named result.txt.

```python
with open("/content/17.txt", "w") as file:
    file.write("Ali, 20, 88\n")
    file.write("Sara, 21, 75\n")
    file.write("Bilal, 22, 90\n")
    file.write("Ayesha, 19, 67\n")
    file.write("Hassan, 23, 82\n")
    file.write("Fatima, 20, 79\n")
    file.write("Usman, 22, 95\n")
    file.write("Noor, 21, 60\n")
    file.write("Zain, 20, 85\n")
    file.write("Maryam, 23, 70\n")
results = {}

with open("/content/17.txt", "r") as file:
    for line in file:
        name, age, grade = line.strip().split(",")
        name = name.strip()
        grade = int(grade.strip())

        if grade >= 80:
            category = "Above Average"
        else:
            category = "Below Average"

        results[name] = category

with open("/content/c2.txt", "w") as output:
    for name, category in results.items():
        output.write(f"{name}: {category}\n")

print("Student Performance Summary:")
for name, category in results.items():
    print(f"{name}: {category}")
```

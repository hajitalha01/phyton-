```python
import random

# Maze size
size = 5

# Player aur treasure ki random position
player_x = 0
player_y = 0
treasure_x = random.randint(0, size-1)
treasure_y = random.randint(0, size-1)

print("🏴‍☠️ Welcome to Maze Adventure!")
print("Maze size 5x5 hai.")
print("Move karo: up, down, left, right")
print("Treasure dhundo!\n")

while True:
    print(f"\n📍 Tumhari position: ({player_x}, {player_y})")
    move = input("Move likho: ").lower()

    if move == "up" and player_x > 0:
        player_x -= 1
    elif move == "down" and player_x < size-1:
        player_x += 1
    elif move == "left" and player_y > 0:
        player_y -= 1
    elif move == "right" and player_y < size-1:
        player_y += 1
    else:
        print("❌ Galat move ya boundary cross!")
        continue

    if player_x == treasure_x and player_y == treasure_y:
        print("💰 Mubarak ho! Tumhe treasure mil gaya!")
        break
```

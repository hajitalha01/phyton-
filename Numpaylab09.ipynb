3. Extract the diagonal elements of a 4x4 identity matrix.

```python
import numpy as np
I = np.eye(4)
print("\n ", I)
diag_elements = np.diag(I)

print("\n \n Diagonal elements:\n" ,diag_elements)
```

4. Create a 2D array of shape (5, 5) with random integers between 0 and 9, and replace all even
values with -1.

```python
import numpy as np
arr = np.random.randint(0, 10, size=(5, 5))

arr[arr % 2 == 0] = -1

print(arr)
```

5. Given two arrays a and b, find the common elements between them.

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([3, 4, 5, 6, 7])

common_elements = np.intersect1d(a, b)

print(common_elements)
```

6. Calculate the matrix product of a 3x4 matrix and a 4x2 matrix.

```python
import numpy as np

A = np.random.randint(1, 10, size=(3, 4))

B = np.random.randint(1, 10, size=(4, 2))

C = A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Product C:\n", C)
```

10. Generate an array of 10 random integers between 50 and 150 using Numpy. Find the minimum,
maximum, and the sum of the elements in the array.

```python
import numpy as np

arr = np.random.randint(50, 151, size=10)

min_val = arr.min()

max_val = arr.max()

sum_val = arr.sum()

print("Array:", arr)
print("Minimum:", min_val)
print("Maximum:", max_val)
print("Sum:", sum_val)
```

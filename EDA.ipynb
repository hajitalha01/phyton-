Task 1: Use the dataset given below of grades of students in a class. Calculate the frequency of each
grade and then create a histogram and a bar chart to visualize distribution of grades among the
students.
Dataset: A, B, C, B, A, C, C, B, A, A, B.

```python
import matplotlib.pyplot as plt
from collections import Counter

grades = ['A', 'B', 'C', 'B', 'A', 'C', 'C', 'B', 'A', 'A', 'B']

frequency = Counter(grades)
print("Frequency of grades:", frequency)

plt.figure()
plt.bar(frequency.keys(), frequency.values())
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.title("Bar Chart of Grades Distribution")
plt.show()

grade_mapping = {'C': 1, 'B': 2, 'A': 3}
numeric_grades = [grade_mapping[g] for g in grades]

plt.figure()
plt.hist(numeric_grades)
plt.xlabel("Grades (C=1, B=2, A=3)")
plt.ylabel("Frequency")
plt.title("Histogram of Grades Distribution")
plt.show()
```

Task 2: Find the frequency of alphabets from the text below and visualize it using a bar chart.
Text: Airports are always a bustling entity with many emotions all over the place. Some are happy to
welcome their loved ones in their arms, some are sad to see them go, while others are anxiously waiting
for their flights to reach. Rolling suitcases, hurrying footsteps, laughing children and the electronic ding
of the machines, all accumulated in one place.

```python
import matplotlib.pyplot as plt
from collections import Counter

text = """
Airports are always a bustling entity with many emotions all over the place.
Some are happy to welcome their loved ones in their arms, some are sad to see them go,
while others are anxiously waiting for their flights to reach.
Rolling suitcases, hurrying footsteps, laughing children and the electronic ding
of the machines, all accumulated in one place.
"""

clean_text = [ch for ch in text.lower() if ch.isalpha()]

frequency = Counter(clean_text)
print("Alphabet Frequency:")
print(frequency)

alphabets = sorted(frequency.keys())
counts = [frequency[a] for a in alphabets]

plt.figure()
plt.bar(alphabets, counts)
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.title("Frequency of Alphabets in Given Text")
plt.show()
```

Task 3: Find the mean, median, mode, variance, range, variance, standard deviation, quartiles (Q1, Q2,
Q3, IQR) and box plot of the data given below.
Data: 23, 45, 56, 34, 78, 90, 12, 67, 43, 55, 88, 32, 67, 54, 78, 89, 22, 35, 66, 43.
(Use the library of numpy, statistics and matplotlib for this question.
Formula for quartiles: np.percentile (data, 25/50/75).
Formula for IQR: q3-q1 Formula for boxplot:
plt.boxplot(data)
Plt.show()

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

data = [23, 45, 56, 34, 78, 90, 12, 67, 43, 55,
        88, 32, 67, 54, 78, 89, 22, 35, 66, 43]

mean_value = np.mean(data)

median_value = np.median(data)

mode_value = stats.mode(data)

variance_value = np.var(data)

std_value = np.std(data)

range_value = max(data) - min(data)


q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

iqr = q3 - q1


print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_value)
print("Range:", range_value)
print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
print("IQR:", iqr)


plt.boxplot(data)
plt.title("Box Plot of Given Data")
plt.show()
```

Task 4: Use the dataset of Laptops Price Dataset and solve the following questions.
1. Print the header, information and describe the data.
2. Find the number of brands in the dataset, display their counts and visualize them in a bar chart.
3. Find the mean, median, mode and variance of the price of laptops and visualize it in a box plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/content/laptops.csv")

print("Header of dataset:")
print(df.head())


print("\nInformation about dataset:")
print(df.info())

print("\nDescription of numerical data:")
print(df.describe())
```

```python

brand_counts = df['Brand'].value_counts()
print("\nNumber of brands:", df['Brand'].nunique())
print("\nCounts of each brand:")
print(brand_counts)

plt.figure(figsize=(12,6))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Counts of Each Laptop Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

```python

mean_price = df['Final Price'].mean()
median_price = df['Final Price'].median()
mode_price = df['Final Price'].mode()[0]
variance_price = df['Final Price'].var()

print("\nLaptop Price Statistics:")
print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Mode Price:", mode_price)
print("Variance of Price:", variance_price)


plt.figure(figsize=(8,6))
sns.boxplot(y=df['Final Price'])
plt.title("Box Plot of Laptop Prices")
plt.ylabel("Price")
plt.show()
```

Question 9: [Marks 5]
Write a Python program that takes a dictionary of students and their marks as input (e.g., {'Ali': 85,
'Sara': 92, 'Bilal': 45}) and:
• Uses if-elif-else to categorize each student as ‘Pass with Distinction’, ‘Pass’, or ‘Fail’.
• Store the categorized results in a new dictionary.
• Finally, display the total count of students in each category.

```python

students = {'Ali': 85, 'Sara': 92, 'Bilal': 45, 'Hina': 78, 'Omar': 60}

result = {}

for name, marks in students.items():
    if marks >= 80:
        result[name] = 'Pass with Distinction'
    elif marks >= 50:
        result[name] = 'Pass'
    else:
        result[name] = 'Fail'

print("Student Results:")
for name, category in result.items():
    print(f"{name}: {category}")

distinction_count = list(result.values()).count('Pass with Distinction')
pass_count = list(result.values()).count('Pass')
fail_count = list(result.values()).count('Fail')

print("\nCategory Summary:")
print("Pass with Distinction:", distinction_count)
print("Pass:", pass_count)
print("Fail:", fail_count)
```

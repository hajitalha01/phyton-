```python
from abc import ABC, abstractmethod

class UniversityManagementSystem(ABC):
    def __init__(self, name, age):
        self.name = name
        self.age = age

  
    def display_role(self):
        pass


class StaffMember:
    def __init__(self, salary):
        self._salary = salary

    def get_salary(self):
        return self._salary

    def set_salary(self, salary):
        if salary < 0:
            print("Not allowed")
        else:
            self._salary = salary


class Student(UniversityManagementSystem):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def display_role(self):
        print("I am a student. My ID is", self.student_id)


class Teacher(UniversityManagementSystem):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def display_role(self):
        print("I am a teacher. My subject is", self.subject)


def main():
    s = Student("Ali", 20, 1)
    t = Teacher("Ahmed", 35, "Math")
    s.display_role()
    t.display_role()


main()
```

```python
from collections import deque
jug1, jug2 = 4, 3

# goal: 2 liters in 4-liter jug
goal = 2

# Function to get all possible next states
def next_states(x, y):
    states = []
    # Fill 4L jug
    states.append((jug1, y))
    # Fill 3L jug
    states.append((x, jug2))
    # Empty 4L jug
    states.append((0, y))
    # Empty 3L jug
    states.append((x, 0))
    # Pour 4 -> 3
    pour = min(x, jug2 - y)
    states.append((x - pour, y + pour))
    # Pour 3 -> 4
    pour = min(y, jug1 - x)
    states.append((x + pour, y - pour))
    return states

# -------- BFS Algorithm --------
def bfs():
    queue = deque()
    queue.append([(0, 0)])
    visited = set()

    while queue:
        path = queue.popleft()
        x, y = path[-1]

        if x == goal:
            return path  # goal found

        for (nx, ny) in next_states(x, y):
            if (nx, ny) not in visited:
                visited.add((nx, ny))
                new_path = list(path)
                new_path.append((nx, ny))
                queue.append(new_path)
    return None

# -------- DFS Algorithm --------
def dfs():
    stack = [[(0, 0)]]
    visited = set()

    while stack:
        path = stack.pop()
        x, y = path[-1]

        if x == goal:
            return path

        for (nx, ny) in next_states(x, y):
            if (nx, ny) not in visited:
                visited.add((nx, ny))
                new_path = list(path)
                new_path.append((nx, ny))
                stack.append(new_path)
    return None

# Run both searches
print("BFS Solution Path:", bfs())
print("DFS Solution Path:", dfs())
```

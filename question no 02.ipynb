```python
import random
import statistics
menu = {
    "Biryani": 350,
    "Karahi": 1200,
    "Burger": 250,
    "Pizza": 800,
    "Fries": 150,
    "Pasta": 500,
    "Sandwich": 200,
    "BBQ": 600,
    "Nihari": 400,
    "Ice Cream": 180
}

customer_bills = []

num_customers = int(input("Enter number of customers (minimum 2): "))

for customer in range(1, num_customers + 1):
    print(f"\n--- Customer {customer} Menu ---")
    items = list(menu.items())
    random.shuffle(items)

    for i, (dish, price) in enumerate(items, 1):
        print(f"{i}. {dish} - Rs.{price}")

    total_bill = 0

    while True:
        choice = input("\nEnter dish number to order (or 'done' to finish): ")

        if choice.lower() == "done":
            break

        choice = int(choice)
        dish_price = items[choice - 1][1]
        total_bill += dish_price
        print(f"Added Rs.{dish_price}")

    customer_bills.append(total_bill)
    print(f"Customer {customer} Bill: Rs.{total_bill}")

print("\n====== BILL SUMMARY ======")
for i, bill in enumerate(customer_bills, 1):
    print(f"Customer {i} Final Bill: Rs.{bill}")

print("\nMaximum Bill:", max(customer_bills))
print("Minimum Bill:", min(customer_bills))
print("Mean Bill:", statistics.mean(customer_bills))
print("Median Bill:", statistics.median(customer_bills))
print("Total Restaurant Earnings:", sum(customer_bills))do
```

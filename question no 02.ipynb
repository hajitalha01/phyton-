Write a recursive function sum_nested_list(lst) that calculates the sum of all numbers in a possibly
nested list, e.g.:
lst = [1, [2, [3, 4], 5], 6]

```python
def sum_nested_list(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total

numbers = [1, [2, [3, 4], 5], 6]
print("Sum of nested list:", sum_nested_list(numbers))
```

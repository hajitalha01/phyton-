Given a list of strings, write a program to:
• Use a for loop to reverse each string only if its length is even.
• Store the modified strings in a tuple.
• Print both the original and final tuple.

```python
words = ["apple", "banana", "kiwi", "pear", "mango", "grapes"]

original_tuple = tuple(words)

modified = []


for word in original_tuple:
    if len(word) % 2 == 0:
        modified.append(word[::-1])
    else:
        modified.append(word)

modified_tuple = tuple(modified)


print("Original Tuple:", original_tuple)
print("Modified Tuple:", modified_tuple)
```
